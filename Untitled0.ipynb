```python
import seaborn as sns
```

```python
plt.figure(figsize=(10,5))
sns.boxplot(data=df[["Age","Fare"]],palette="dark")
plt.title("boxplot for outlier detection")
plt.show()
```

```python
plt.figure(figsize=(10,15))
sns.countplot(x="Survived",data=df,palette="deep")
plt.title("survival count",fontsize=15)
plt.show()
```

```python
plt.figure(figsize=(20,15))
sns.histplot(df["Age"],bins=30,kde=True,color="red")
plt.title("Age distribution",fontsize=15)
plt.show()
```

```python
plt.figure(figsize=(10,15))
sns.scatterplot(data=df[["Age","Fare"]],palette="pastel")
plt.title("scatter plot",fontsize=20)
plt.show()
```

```python
plt.figure(figsize=(10,5))
sns.violinplot(x="Survived",y="Fare",data=df,palette="muted")
plt.title("violin plot")
plt.show()
```

```python
plt.figure(figsize=(10,5))
sns.kdeplot(df["Age"],shade=True,color="blue")
plt.title("density plot")
plt.show()
```
