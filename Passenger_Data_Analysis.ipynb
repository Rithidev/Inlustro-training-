```python
duplicate_rows_titanic=titanic[titanic.duplicated()]
print("number of duplicate rows:",duplicate_rows_titanic.shape)
```

```python
print(titanic.isnull().sum())
```

```python
titanic = titanic.fillna({'Age': titanic['Age'].median()})
```

```python
titanic.dropna()
```

```python
plt.figure(figsize=(10,5))
sns.boxplot(data=titanic[["Age","Fare"]],palette="dark")
plt.title("boxplot for outlier detection")
plt.show()
```

```python
plt.figure(figsize=(10,5))
sns.histplot(titanic["Age"],bins=30,kde=True,color="purple")
plt.title("Age Distribution",fontsize=14)
plt.show()
```

```python
plt.figure(figsize=(10, 8))
sns.violinplot(x="Sex", y="Age", data=titanic, hue="Survived", split=True)
plt.title("Age Distribution by Sex", fontsize=14)
plt.show()
```

```python
plt.figure(figsize=(10,5))
sns.barplot(x="Survived", y="Fare", data=titanic, hue="Sex", palette="Blues")
plt.title("Fare Distribution by Survival", fontsize=14)
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.countplot(x="Survived", data=titanic, hue="Sex", palette="deep")
plt.title("Survival Count", fontsize=15)
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(x="Sex", y="Fare", data=titanic, hue="Survived", palette="viridis")
plt.title("Fare vs Family Size by Survival", fontsize=14)
plt.show()
```
