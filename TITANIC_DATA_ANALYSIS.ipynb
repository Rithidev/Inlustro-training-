```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.preprocessing import LabelEncoder, OneHotEncoder
# Assuming 'titanic' is your DataFrame
titanic=pd.read_csv("/content/train (1).csv")

# Drop unnecessary columns
titanic = titanic.drop(["SibSp", "Embarked", "Cabin"], axis=1)

# Fill missing values in 'Age' with the median
titanic = titanic.fillna({'Age': titanic['Age'].median()})

# Drop rows with any remaining missing values
titanic = titanic.dropna()

# Separate features and target variable
X = titanic.drop('Survived', axis=1)  # Features
y = titanic['Survived']  # Target

# Create a LabelEncoder for the 'Sex' column
le = LabelEncoder()
X['Sex'] = le.fit_transform(X['Sex'])
he = OneHotEncoder(sparse_output=False, handle_unknown='ignore')
name_encoded = ohe.fit_transform(X[['Name']])

# Handle the 'Name' column (example using OneHotEncoder)
# 1. Create a OneHotEncoder
ohe = OneHotEncoder(sparse_output=False, handle_unknown='ignore')
# 2. Fit and transform the 'Name' column
name_encoded = ohe.fit_transform(X[['Name']])
# 3. Create a DataFrame from the encoded features
name_df = pd.DataFrame(name_encoded, columns=ohe.get_feature_names_out(['Name']))
# 4. Drop the original 'Name' column and concatenate the encoded features
X = X.drop('Name', axis=1)
X = pd.concat([X, name_df], axis=1)
# 5.Handle Ticket column (You can either drop or use similar approach as Name)
X = X.drop('Ticket',axis=1) #dropping for simplicity

# Split data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Create and train the model
clf = RandomForestClassifier(n_estimators=100, random_state=42)
clf.fit(X_train, y_train)
```

```python
y_pred=clf.predict(X_test)
```

```python
accuracy=accuracy_score(y_test,y_pred)
print("Accuracy:",accuracy)
```
